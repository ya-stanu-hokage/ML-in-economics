Бизнес метрика (metric) будет формироваться из четырех составных фичей:
 - (F1) человеческий ресурс ('RESEARCHERS', 'STUDENT_GRAD') - чем больше народу, тем больше идей
 - (F2) уровень инновационной активности организаций ('SHARE_IA_FIRMS') - без заинтересованной организации сложно что-то сделать масштабно и качественно
 - (F3) патенты ('PATENT_ISSUED_INV', 'PATENT_ISSUED_MOD') - патенты-хороший показатель активности исследований
 - (F4) денежный ресурс ('FRIE', 'ARIE', 'DIE') - без денег ничего не реализуешь

F1 = 'RESEARCHERS' + 1000 * 0.02 * 'STUDENT_GRAD' (учитываем, что лишь 2% от числа выпускников собирается заниматься чем-то инновационными (а не просто косят от армии или ждут мужа))

F2 = 'SHARE_IA_FIRMS' (уровень инновационной активности организаций)

F3 = 'PATENT_ISSUED_INV' + 'PATENT_ISSUED_MOD' (общее число патентов)

F4 = 'FRIE' + 'ARIE' + 'DIE' (учитваем сколько денег готовы вливать в инновации организации)

Все фичи нормированы через MaxMinScaler [0, 1]

metric = 0.6 * F1 + 0.12 * F2 + 0.03 * F3 + 0.25 * F4 (веса с неба, как чувствую) [0, 1]

```python
df21 = df21[['Region',
             'RESEARCHERS', 'STUDENT_GRAD', #человеческий ресурс
             'SHARE_IA_FIRMS', #уровень инновационной активности организаций
             'PATENT_ISSUED_INV', 'PATENT_ISSUED_MOD', #патенты
            'FRIE', 'ARIE', 'DIE']] #денежный ресурс
df21
```

```python
df21['humans_feature'] = df21['RESEARCHERS'] + 1000 * 0.02 *df21['STUDENT_GRAD']
df21['businesses_feature'] = df21['SHARE_IA_FIRMS']
df21['patents_feature'] = df21['PATENT_ISSUED_INV'] + df21['PATENT_ISSUED_MOD']
df21['money_feature'] = df21['FRIE'] + df21['ARIE'] + df21['DIE']
df21 = df21[['Region', 'humans_feature', 'businesses_feature', 'patents_feature', 'money_feature']]
df21.head()
```

```python
ind = df21['Region'].str.contains('округ')
df21 = df21[~ind]
df21 = df21[~df21['Region'].str.contains('Россия')]
```

Попробуем расчитать вначале метрику с учетом данных субъектов

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sc = MinMaxScaler()
df21_sc = df21.copy()
df21_sc.iloc[:, 1:] = sc.fit_transform(df21_sc.iloc[:, 1:])
df21_sc
```

```python
df21_sc['metric'] = 0.6 * df21_sc['humans_feature'] + 0.12 * df21_sc['businesses_feature'] + 0.03 * df21_sc['patents_feature'] + 0.25 * df21_sc['money_feature']
```

```python
df21_sc.sort_values(by='metric', ascending=False).reset_index()
```

Выведем топ 15 субъетов РФ по выбранной метрике

```python
top15 = df21_sc.sort_values(by='metric', ascending=False)[['Region', 'metric']].head(15)
plt.figure(figsize=(15,8))
plt.title('Топ субъектов по инновационному потенциалу', fontdict={'size':20})
plt.bar(height=top15['metric'], x=top15['Region'])
plt.gca().set_xticklabels(top15['Region'], rotation=60, horizontalalignment= 'right')
plt.grid()
plt.show()
```

Интерсно посмотреть, на метрику без учета этих 4 субъектов:

```python
subs = ['Моск', 'г.Санкт', 'Нижегор']
df21_wo_sc = df21.copy()
for el in subs:
    df21_wo_sc = df21_wo_sc[~df21_wo_sc['Region'].str.contains(el)]

sc_wo = MinMaxScaler()
df21_wo_sc.iloc[:, 1:] = sc_wo.fit_transform(df21_wo_sc.iloc[:, 1:])

df21_wo_sc['metric'] = 0.6 * df21_wo_sc['humans_feature'] + 0.12 * df21_wo_sc['businesses_feature'] + 0.03 * df21_wo_sc['patents_feature'] + 0.25 * df21_wo_sc['money_feature']
df21_wo_sc.sort_values(by='metric', ascending=False).reset_index()
```

```python
top15_wo = df21_wo_sc.sort_values(by='metric', ascending=False)[['Region', 'metric']].head(15)
plt.figure(figsize=(15,8))
plt.title('Топ субъектов по инновационному потенциалу', fontdict={'size':20})
plt.bar(height=top15_wo['metric'], x=top15_wo['Region'])
plt.gca().set_xticklabels(top15_wo['Region'], rotation=60, horizontalalignment= 'right')
plt.grid()
plt.show()
```

```python
print(* list(set(top15.Region) & set(top15_wo.Region)), sep=', ')
print(len(set(top15.Region) & set(top15_wo.Region)))
```

Данные топы сильно пересекаются (11 из топ 15), хоть и ранжирование немного поменялось, но не некритично.

Выбранная метрика расчета инновационнго потенциала неплохо отражает реальность. Это случилось из-за того, что многонаселенные и богатые субъекты находятся в топе, так как фичи человеского ресурса и денежного обладают большими весами в метрике. Но все равно данная метрика взята с неба и по моему наитию. Вроде наитие работает неплохо.
